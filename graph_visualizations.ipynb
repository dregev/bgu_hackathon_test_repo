```python
import networkx as nx
```

```python
G=nx.Graph()
G.add_nodes_from(range(10))
```

```python
G.add_edges_from([(1,2),(1,3),(1,6),(2,3),(4,7),(4,9),(3,7),(8,9)])
```

```python
print(G.nodes)
print(G.edges)
```

```python
list(nx.connected_components(G))
```

```python
sorted(d for n, d in G.degree())
```

```python
nx.clustering(G)
```

```python
import matplotlib.pyplot as plt
```
